```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data for the algorithms
fedavg = [0.9800, 0.9806, 0.9801, 0.9800]
fedprox = [0.9680, 0.9686, 0.9652, 0.9663]

# Create a 3D bar chart
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')

# X positions for metrics
x_pos = np.arange(4)  # Four metrics: Accuracy, Precision, Recall, F1-Score

# Create a grid for the bars
y_pos = [0, 1]  # Two bars for each metric (FedAvg and FedProx)
x_pos_grid, y_pos_grid = np.meshgrid(x_pos, y_pos)

# Flatten the positions
x_pos_grid = x_pos_grid.flatten()
y_pos_grid = y_pos_grid.flatten()

# Heights for FedAvg and FedProx
heights = fedavg + fedprox

# Professional color scheme
fedavg_color = '#1f77b4'  # Blue
fedprox_color = '#2ca02c'  # Green
colors = [fedavg_color] * 4 + [fedprox_color] * 4

# Create 3D bars
ax.bar3d(x_pos_grid, y_pos_grid, np.zeros_like(x_pos_grid), 0.4, 0.4, heights, color=colors, shade=True)

# Set labels for axes with slightly closer padding for 'Algorithms'
ax.set_xlabel('Metrics', fontsize=14, fontweight='bold', color='black', labelpad=20)  # Dark font color
ax.set_ylabel('Algorithms', fontsize=14, fontweight='bold', color='black', labelpad=15)  # Dark font color
ax.set_zlabel('Performance', fontsize=14, fontweight='bold', color='black', labelpad=15)  # Dark font color

# Set ticks for x and y axes with larger spacing and moved farther from the bars
ax.set_xticks(x_pos)
ax.set_xticklabels(['Accuracy', 'Precision', 'Recall', 'F1-Score'], fontsize=12, fontweight='normal', color='black')
ax.set_yticks([0, 1])
ax.set_yticklabels(['FedAvg', 'FedProx'], fontsize=12, fontweight='normal', color='black')

# Title with smaller font size
ax.set_title('Performance Comparison: FedAvg vs FedProx Algorithms', fontsize=16, fontweight='bold', color='black')

# Annotate the performance values above the bars with smaller font size
for i in range(len(fedavg)):
    ax.text(x_pos[i] - 0.05, 0, fedavg[i] + 0.02, f'{fedavg[i]*100:.2f}%', color='black', ha='center', va='bottom', fontsize=10)
    ax.text(x_pos[i] + 0.05, 1, fedprox[i] + 0.02, f'{fedprox[i]*100:.2f}%', color='black', ha='center', va='bottom', fontsize=10)

# Customize the appearance
ax.view_init(elev=30, azim=150)  # Change the view angle for a dynamic look

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
epsilon_values = [2.0, 1.5, 1.0]
delta_values = [1e-5, 1e-5, 1e-6]
sensitivity_values = [0.05, 0.05, 0.06]
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']

# Metric values
results = [
    [0.950, 0.956, 0.950, 0.950],  # epsilon = 2.0, delta=1e-5, sensitivity=0.05
    [0.932, 0.9294, 0.9313, 0.9299],  # epsilon = 1.5, delta=1e-5, sensitivity=0.05
    [0.9096, 0.9058, 0.9086, 0.9063]  # epsilon = 1.0, delta=1e-6, sensitivity=0.06
]

# Convert to numpy array
data = np.array(results)

# Plot setup
plt.figure(figsize=(10, 6))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Professional color palette
markers = ['o', 's', '^', 'd']  # Added 4th marker 'd' (diamond shape)

# Plot lines for each metric
for i, metric in enumerate(metrics):
    plt.plot(epsilon_values, data[:, i], marker=markers[i],
             linestyle='-', linewidth=2, markersize=8, label=metric, color=colors[i])

# Add annotations for delta and sensitivity
for i, epsilon in enumerate(epsilon_values):
    annotation_text = f"={delta_values[i]}, S={sensitivity_values[i]}"
    plt.text(epsilon, data[i, 0] - 0.002, annotation_text, fontsize=10,
             ha='center', va='bottom', color='black', bbox=dict(facecolor='white', alpha=0.7, edgecolor='gray'))

# Labels and Titles
plt.title('Performance Metrics with Epsilon, Delta, and Sensitivity', fontsize=14, fontweight='bold')
plt.xlabel('Epsilon Values', fontsize=12, fontweight='bold')
plt.ylabel('Metric Value', fontsize=12, fontweight='bold')

# Axis settings
plt.xticks(epsilon_values, fontsize=10)
plt.yticks(np.arange(0.90, 0.97, 0.01), fontsize=10)
plt.ylim(0.90, 0.97)

# Add grid and legend at the top
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title="Metrics", loc='upper center', fontsize=10, bbox_to_anchor=(0.5, 1.25), ncol=4)

# Adjust layout and show
plt.tight_layout()
plt.show()

# Limitation Text:
limitation_text = (
    "Limitations include low performance of FL, binary classification, traditional augmentation, "
    "and lack of differential privacy mechanisms."
)

print(limitation_text)
```
